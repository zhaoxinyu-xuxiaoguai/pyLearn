# 聚合函数

```python
n=np.random.randint(20,size=(3,4))
sum0=np.sum(n,axis=0)#每一行相加
sum1=np.sum(n,axis=1)#每一列相加
display(n,sum0,sum1)
```

```python
#常见聚合函数
"""
np.min 最小值
np.max
np.mean
np.average或np.mean
np.median中位数
np.argmin最小值对应的下标
np.argmax最大值对应下标
np.std标准差
np.var方差
np.power次方，求幂
np.argwhere 按条件查找
"""
n=n.reshape(-1)#将n展平为一维数组
print(n)
np.argmax(n)
```

```python
# 矩阵操作
n=np.random.randint(0,10,size=(3,5))
print(n)
```

```python
n+10#矩阵所有元素都进行运算，+，-，*，/，//，**同理
n1=np.random.randint(10,size=(3,4))
n2=np.random.randint(10,size=(3,4))
n1+n2 #对应元素相加
```

```python
#np.dot()矩阵乘法，n1*n2是对应元素相乘
n3=np.random.randint(20,size=(4,3))
n4=np.dot(n1,n3)
n4
```

```python
#矩阵的逆
np.linalg.inv(n4)
```

```python
#矩阵行列式
t1=np.round(np.linalg.det(n4))
t2=np.round(np.linalg.det(np.linalg.inv(n4)))
display(t1,t2)
```

# 广播机制

会自动根据已有的值，填充没有的值

```python
m=np.ones((2,3),dtype=np.int8)
n=np.arange(3)
display(m,n)
```

```python
m+n#给n矩阵再补了一行【0,1,2】
```

```python
m=np.arange(3).reshape((3,1))
n=np.arange(3)
display(m,n)
```

```python
m+n
# """
# m会变成：
# 0 0 0
# 1 1 1
# 2 2 2
# n变成：
# 0 1 2
# 0 1 2
# 0 1 2 
# """
```

# numpy文件操作

```python
#save:保存ndarray到一个npy文件
#savez：将多个array保存到一个npz文件
x=np.arange(10)
y=np.arange(10,20)
#save,保存为x.npy
np.save('x',x)

#savez
np.savez('arr',xarr=x,yarr=y)#自定义标签
```

```python
#读取npy文件
np.load('x.npy')
```

```python
#读取npz
np.load('arr.npz')
```

```python
np.load('arr.npz')['xarr']
```

```python
np.load('arr.npz')['yarr']
```
